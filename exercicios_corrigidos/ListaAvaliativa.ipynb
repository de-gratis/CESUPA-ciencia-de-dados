## 5.

## 7.
Descubra os bairros sem ruas (valores NaN) e imprima a quantidade de ruas ausentes.

```python
bairros_sem_rua = df['Bairro'][pd.isnull(df['Rua'])].value_counts()
bairros_sem_rua
```
