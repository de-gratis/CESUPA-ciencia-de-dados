## Trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

## 1. 
Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
df_total = pd.concat([df1, df2], ignore_index=True).drop(
    ['Unnamed: 0'], axis=1)

df_total
```

## 4.
Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método `pd.to_datetime()`, converta ambas colunas para Timestamp / Datetime.

```python
df_datetime = df_total[['DATA INICIAL', 'DATA FINAL']].apply(pd.to_datetime)

df_datetime
```

## 5.
Crie uma nova coluna para representar o mês e o ano (mm-aaaa), utilizando a coluna DATA FINAL como referência. 

```python
df_total['MES/ANO'] = df_datetime['DATA FINAL'].apply(
    lambda x: x.strftime('%m-%Y')).apply(pd.to_datetime)
    
df_total['MES/ANO']
```

## 8.
Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
df_total['PREÇO MÉDIO REVENDA'][
    (df_total['MES/ANO'].dt.year == 2008) &
    (df_total['MES/ANO'].dt.month == 8)]
```

## 9.
Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
df_total['PREÇO MÉDIO REVENDA'][
    (df_total['MES/ANO'].dt.year == 2014) & 
    (df_total['MES/ANO'].dt.month == 5) & 
    (df_total['ESTADO'] == 'SAO PAULO')]
```
