```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Gráfico de Barras; Bar plot

```python
df = pd.DataFrame.from_dict({'Excelente': 58, 
                             'Bom': 18, 
                             'Regular': 32, 
                             'Ruim': 12}, 
                             orient='index', columns=['Frequência'])

plt.bar(df.index, df['Frequência'].values)

plt.xlabel('Satisfação')
plt.ylabel('Frequência')

plt.show()
```

# Gráfico Pizza; Pie chart

```python
valores = {'PMDB': 18, 
           'Rede': 22, 
           'PDT': 12, 
           'PSDB': 25, 
           'PC do B': 8, 
           'PV': 5, 
           'Outros': 10, }

df = pd.DataFrame.from_dict(valores, orient='index', columns=['Valor'])

df.plot.pie(y='Valor', figsize=(6, 6), autopct='%1.1f%%', startangle=170)

plt.show()
```

# Diagrama de Pareto; Pareto chart

```python
data = {'Tipo de defeito': ['Amassado', 'Perfurado', 
                            'Impressão ilegível', 'Caracteres errados', 
                            'Números errados', 'Outros'],
        'Frequência absoluta (Fi)': [71, 28, 
                                     12, 20, 
                                     44, 6]}

df = pd.DataFrame(data)
df = df.sort_values('Frequência absoluta (Fi)', ascending=False)

df['Frequencia_acumulada'] = df['Frequência absoluta (Fi)'].cumsum()

total = df['Frequência absoluta (Fi)'].sum()
df['Percentual'] = round(100 * df['Frequencia_acumulada'] / total, 2)

sns.set_style("whitegrid")
fig, ax = plt.subplots()
ax.bar(df['Tipo de defeito'], df['Frequência absoluta (Fi)'], color="C0")
ax2 = ax.twinx()
ax2.plot(df['Tipo de defeito'], df['Percentual'], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(plt.FuncFormatter('{:.0f}%'.format))
ax.tick_params(axis='y', labelcolor="C0")
ax2.tick_params(axis='y', labelcolor="C1")
ax.set_ylabel('Frequência', color="C0")
ax2.set_ylabel('Percentual', color="C1")
plt.xticks(rotation=45)
table = plt.table(cellText=df.values, colLabels=df.columns, loc='bottom')

table.auto_set_column_width(col=list(range(len(df.columns))))

table.scale(1, 1.5)

plt.axis('off')

plt.show()

# Por ajeitar
```

# Gráfico de Linhas; Line chart

```python
values = [.42, .38, .12, .34, .22, .15, .18, .31, .47, .24, .42, .09]
dates = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 
         'Maio', 'Junho', 'Julho', 'Agosto', 
         'Setembro', 'Outubro', 'Novembro', 'Dezembro']
         
data = pd.DataFrame(values, dates)
plt.figure(figsize=(14, 4))
plt.ylabel('Perdas (%)')
plt.xlabel('Mês')

sns.lineplot(data=data, palette="tab10", linewidth=1)
```

# Gráfico de Pontos ou de Dispersão; Scatter plot

```python
data = pd.DataFrame({'Tempo (min)': [30, 50, 85, 106, 138, 
                                     178, 198, 222, 252, 297],
                     'Peso (kg)': [250, 255, 252, 248, 250, 
                                   249, 252, 251, 250, 245]})

sns.relplot(x="Tempo (min)", y="Peso (kg)", data=data, height=4, aspect=2)
```
